```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('student-mat.csv', sep=';')
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.drop(columns=['G2','G3'],inplace=True)
df
```

```python
Y=df['G1'].values
Y
```

```python
categorical_cols=df.select_dtypes(include=('object')).columns
numeric_cols=df.select_dtypes(exclude=('object')).columns
```

```python
new_df=pd.get_dummies(df,columns=categorical_cols, drop_first=True)
new_df
```

```python
X=new_df.drop(columns=['G1']).values
X
```

```python
from sklearn.model_selection import train_test_split
x_train ,x_test ,y_train , y_test=train_test_split(X,Y,test_size=0.2,random_state=42 )
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
from sklearn.linear_model import Lasso
Lasso_Model=Lasso()
Lasso_Model.fit(x_train, y_train)
```

```python
preds=Lasso_Model.predict(x_test)
print(f"w0:{Lasso_Model.intercept_}")
print(f"w1:{Lasso_Model.coef_}")

```

```python
from sklearn.metrics import mean_squared_error,r2_score,root_mean_squared_error

print("Mean Squared Error:", mean_squared_error(y_test,preds))
print("Root Mean Squared Error:", root_mean_squared_error(y_test,preds))
print("R² Score:", r2_score(y_test,preds))
```
