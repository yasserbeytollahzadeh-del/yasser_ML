```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import root_mean_squared_error ,  mean_absolute_error , mean_squared_error
```

```python
df=pd.read_csv('FuelConsumption.csv')
df.head()
```

```python
column_set=[
    ['FUELCONSUMPTION_COMB_MPG'],
    ['FUELCONSUMPTION_COMB'],
    ['FUELCONSUMPTION_HWY'],
    ['FUELCONSUMPTION_COMB_MPG','FUELCONSUMPTION_COMB'],
    ['FUELCONSUMPTION_COMB_MPG','FUELCONSUMPTION_HWY'],
    ['FUELCONSUMPTION_COMB','FUELCONSUMPTION_HWY'],
    ['FUELCONSUMPTION_COMB_MPG','FUELCONSUMPTION_COMB','FUELCONSUMPTION_HWY']
]
len(column_set)
```

```python
metrics=[]
for item in column_set:
    x=df[item]
    y=df['CO2EMISSIONS']
    x_train , x_test , y_train , y_test=train_test_split(x,y,test_size=0.2)
    regression= LinearRegression()
    regression.fit(x_train, y_train)
    preds=regression.predict(x_test)
    erroe_rmse=root_mean_squared_error(y_test, preds)
    error_mae=mean_absolute_error(y_test, preds)
    error_mse=mean_squared_error(y_test, preds)
    print(f"column:{item}:ðŸ“ˆ RMSE={erroe_rmse:.4f} â€¢ ðŸ“‰ MAE={error_mae:.4f} â€¢ ðŸ§® MSE={error_mse:.4f}")
    metrics.append((item,erroe_rmse,error_mae,error_mse))
error_df=pd.DataFrame(metrics,columns=['column','root_mean_squared_error','mean_absolute_error','mean_squared_error'])  
error_df  
```

```python
error_df.sort_values(by='mean_absolute_error',ascending=True)
```
