```python
import numpy as np

def Entropy(A,B):
    total=A+B
    p1=A/total
    p2=B/total
    Res=[]
    if p1>0:
        Res.append(-p1*np.log(p1))
    if p2>0:
        Res.append(-p2*np.log(p2))
    return sum(Res)
```

```python
print(Entropy(4,3)) 
print(Entropy(1,6)) 
```

```python
def informationGain(E_old,E_new):
    return E_old-E_new
```

```python
main_Entropy=Entropy(5,9)
A_left_Entropy=Entropy(4,3)
b_Right_Entropy=Entropy(1,6)
total=14
E_new=(7/14)*A_left_Entropy+(7/14)*b_Right_Entropy
IG = informationGain(main_Entropy, E_new)
print("Information Gain:", IG)

```
